#### Importación de bibliotecas en Python:

```python
import os
import matplotlib.pyplot as plt
import librosa, librosa.display
import IPython.display as ipd
import pandas as pd
```

#### Parametrización de ubicaciones de los archivos:

```python
BASE_FOLDER = "audio-dataset/"
diego_sound_file = "Diego.wav"
rodrigol_sound_file = "RodrigoL.wav"
```

#### Cargar los audios para procesamiento:

```python
diego_sound, _ = librosa.load(os.path.join(BASE_FOLDER, diego_sound_file))
rodrigol_sound, _ = librosa.load(os.path.join(BASE_FOLDER, rodrigol_sound_file))
```

#### Función de Gráfica de espectrograma y almacenamiento de dataframe(df):

```python
def plot_spectrogram(signal, name):
    espectrograma = librosa.amplitude_to_db(librosa.stft(signal))
    df = pd.DataFrame(espectrograma)
    df.to_csv(f'espectrograma{name}.csv', index=False)
    plt.figure(figsize=(15, 10))
    librosa.display.specshow(espectrograma, y_axis="log")
    plt.colorbar(format="%+2.0f dB")
    plt.title(f"Espectrograma en log-frecuencia de {name}")
    plt.xlabel("Time")
    plt.show()
```

#### Audio a graficar (ejemplo de Rodrigo Lobos):

```python
ipd.Audio(os.path.join(BASE_FOLDER, rodrigol_sound_file)) 
```

```python

plot_spectrogram(rodrigol_sound, "Rodrigo Lobos")
```
